```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This is just a sample from a big database
eeg_id = 568657
spec_id = 789577333

eeg_sampling_rate = 200 # Hz
eeg_time_period = 50 # seconds
spec_sampling_rate = 0.5 # Hz
spec_time_period = 600 # seconds

eeg_filename = f"eeg_file\\{eeg_id}.parquet"
spec_filename = f"spectogram_file\\{spec_id}.parquet"

df_eeg = pd.read_parquet(eeg_filename)
df_spec = pd.read_parquet(spec_filename)
```

```python
# offset = [0, 6, 12, 16]
# In this specific data point, the offset 
# for both is same, so we are considering single offset
offset = 0
eeg = df_eeg.iloc[offset*eeg_sampling_rate:(offset+eeg_time_period)*eeg_sampling_rate]
spectrogram = df_spec.loc[(df_spec.time>=offset)
                &(df_spec.time<offset+spec_time_period)]
```

```python
# Extract EEG and EKG channels
eeg_channels = eeg.iloc[:, :19]
ekg_channel = eeg['EKG']
```

Computing the FFT of the data

```python
#FFT of EEG and EKG samples
def plot_FFT(channel, fs):
    fft_result = np.fft.fft(channel)
    freq_bins = np.fft.fftfreq(len(channel), d=1 / fs)
    plt.plot(abs(freq_bins), np.abs(fft_result))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.grid(True)



idx1 = 0
plt.figure(figsize=(10,10))
for col in eeg_channels.columns:
    idx1 += 1
    plt.subplot(5, 4, idx1)
    plot_FFT(eeg_channels[col], eeg_sampling_rate)
    plt.title(col)

plt.subplot(5, 4, 20)
plot_FFT(ekg_channel, eeg_sampling_rate)
plt.title(col)

plt.tight_layout()
plt.show()
```

Spectogram Comparision

```python
# Creating the EEG Spectogram from EEG data sample
from scipy.signal import spectrogram

plt.figure(figsize=(10,10))
idx_ = 0
for col in eeg_channels.columns:
    # f --> array of sample frequencies
    # t --> array of segment times
    # Pxx --> Spectogram of x
    f, t, Pxx = spectrogram(eeg_channels[col], fs=200)
    idx_+=1
    plt.subplot(5,4, idx_)
    plt.pcolormesh(t, f, 10 * np.log10(Pxx), cmap='viridis')
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    plt.title(col)
    plt.grid(True)
    
plt.subplot(5,4, 20)
f, t, Pxx = spectrogram(ekg_channel, fs=200)
plt.pcolormesh(t, f, 10 * np.log10(Pxx), cmap='viridis')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('EKG Signal')
plt.grid(True)

plt.tight_layout()  
plt.show()
```

```python
# Plotting the actual spectogram available from the database
# To compare above spectogram

plt.figure(figsize=(12,10))
idx = 0
for col in eeg_channels.columns:
    idx+=1
    plt.subplot(5,4, idx)
    plt.specgram(eeg_channels[col], Fs=eeg_sampling_rate, cmap='viridis')
    plt.colorbar(label='Magnitude')
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    plt.title(col)
    
plt.subplot(5,4, 20)
plt.specgram(ekg_channel, Fs=eeg_sampling_rate, cmap='viridis')
plt.colorbar(label='Magnitude')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('EKG')

plt.tight_layout()  
plt.show()
```
